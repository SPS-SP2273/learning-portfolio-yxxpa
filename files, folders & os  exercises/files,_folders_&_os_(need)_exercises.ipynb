```python
import glob

import os

import shutil
```

**Exercise 1**

**Task 1**

```python
months = {
"jan": "01", "feb": "02", "mar": "03",
"apr": "04", "may": "05", "jun": "06",
"jul": "07", "aug": "08", "sep": "09",
"oct": "10", "nov": "11", "dec": "12"
}
```

```python
filename = "oct-08_singapore.txt"

# Converting the filename
month = months[filename.split("-")[0]]
filename = filename.replace("oct", month).replace("_", "-")

print(filename)
```

**Task 2**

```python
def rename_my_file(old_file_name):
    # Get the current month in the file
    month = old_file_name.split("-")[0]
    # Get the new month
    converted_month = months[month]
    # Convert the current filename and return the new filename
    return old_file_name.replace(month, converted_month).replace("_", "-")
```

**Task 3**

```python
file_list = ["oct-08_singapore.txt", "jul-10_cairo.txt", "may-15_dhaka.txt",
 "may-13_cairo.txt", "oct-21_cairo.txt", "jan-10_singapore.txt",
 "jun-20_tokyo.txt", "aug-06_jakarta.txt", "dec-21_karachi.txt",
 "jan-01_tokyo.txt"]
```

```python
for filename in file_list:
    print(f"{filename} ----> {rename_my_file(filename)}")
```

**Task 4**

```python
# Getting all the files ending in .txt
txt_files = glob.glob("os-collaboration-exercise-data/*.txt")
```

```python
txt_files
```

```python
for i in range(len(txt_files)):
    # Get the name of the file (without the folder name)
    filename = os.path.basename(txt_files[i])
    # Rename the file
    new_filename = rename_my_file(filename)
    new_filename = "os-collaboration-exercise-data/" + new_filename
    os.rename(txt_files[i], new_filename)
```

**Task 5**

```python
cities = ["Cairo", "Dhaka", "Jakarta", "Karachi", "Manila",
 "Melbourne", "Osaka", "Shanghai", "Singapore", "Tokyo"]
```

```python
for city in cities:
    # Create the new folder's path
    path = os.path.join("os-collaboration-exercise-data", city)
    # Create the new folder
    try:
        os.mkdir(path)
        print(f"Creating {path}")
    # If the folder already exists, handle the error.
    except FileExistsError:
        print(f"{path} already exists, so it will be skipped.")
```

**Task 6**

```python
# Listing the files related to Cairo
cairo_files = glob.glob("os-collaboration-exercise-data/*cairo*")
```

```python
cairo_files
```

```python
for i in range(len(cairo_files)):
    # Copying the files related to Cairo into the Cairo folder
    path_to_destination = "os-collaboration-exercise-data/Cairo"
    shutil.copy(cairo_files[i], path_to_destination)
```

**Task 7**

```python
# Listing all the files
all_files = glob.glob("os-collaboration-exercise-data/*")
```

```python
all_files
```

```python
for i in range(len(all_files)):
    for j in range(len(cities)):
        # Getting the filenames with and without the extension
        filename = all_files[i].split("/")[1].split(".")[0]
        filename_with_extension = os.path.basename(all_files[i])
        
        # Getting the current city
        city = cities[j]

        # Getting the path to the current city's folder
        city_path = os.path.join("os-collaboration-exercise-data", city)

        # Getting the current path of the file and
        # writing the desired new path
        current_path = all_files[i]
        new_path = os.path.join(city_path, filename_with_extension)
        
        # If the current file's name ends in the current city, examine it to see
        # whether the file already exists in the corresponding folder.
        # If so, do nothing. If it doesn't exist yet, move the file to the new path.
        if filename.endswith(city.lower()):
            if os.path.exists(new_path):
                pass
            else:
                shutil.move(current_path, new_path)
```
