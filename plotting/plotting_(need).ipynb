# 1 A simple plot

## 1.1 Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label="Y values",
         color="red", linestyle="dashed")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("X vs Y")
plt.grid(alpha=.25)
plt.legend(loc="upper left")
```

### Things to note

```python
# Abbreviations - produces same result
plt.plot(x, y_1, color="red", linestyle="dashed", linewidth=2)
plt.plot(x, y_1, c="r", ls="dashed", lw=2)
```

```python
plt.plot(x, y_1, "r--")
plt.show()  # Good practice to include this line
```

```python
# Plotting multiple lines
plt.plot(x, y_1, label="a", c="k")  
plt.plot(x, y_2, label="b", c="k", ls="--")
plt.legend()
```

```python
# Using default values
plt.plot(x, y_1, y_2)  
```

```python
# Order of keyword arguments does not matter
plt.plot(x, y_1, color="red", linestyle="dashed", linewidth=2)
plt.show()
```

```python
# Order of keyword arguments does not matter
plt.plot(x, y_1, 
         label="Y values",
         color="red", linestyle="dashed")
plt.show()
```

## 1.2 Adding another plot

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label="Y1 values",
         color="red", linestyle="dashed")
plt.plot(x, y_2, label="Y2 values", color="blue")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("X vs Y")
plt.grid(alpha=.25)
plt.legend(loc="upper left")
```

## 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Let's start plotting
plt.plot(x, y_1, label="Y1 values",
         color="red", linestyle="dashed")
plt.plot(x, y_2, label="Y2 values", color="blue")
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error, 
             label="Y3 with errors", color="green", alpha=.5)
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("X vs Y")
plt.grid(alpha=.25)
plt.legend(loc="upper left")
```

# 2 Better with NumPy

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color="red", linestyle="dashed", label="Y values")
plt.plot(np_x, np_y_2, color="blue", label="Y2 values")
```

## 2.1 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                
sin_x = np.sin(x)

plt.plot(x, x2, label="x^2",
         color="lightseagreen", linestyle="dashed")
plt.plot(x, sin_x, label="sin(x)",
         color="orchid")
plt.legend()

# Note: the above will not work with other packages like math.
```

## We need another axis!

### Things to note

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label="x^2",
         color="lightseagreen", linestyle="dashed")
plt.legend(loc="lower left")  # For y-axis 1

plt.twinx()  # This creates a new y-axis 
             # for the plots that comes after

plt.plot(x, sin_x, label="sin(x)",
         color="orchid")
plt.legend(loc="lower right")  # For y-axis 2
```

```python
# Getting a smooth curve
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label="x^2",
         color="lightseagreen", linestyle="dashed")
plt.legend(loc=(.25, .5), frameon=False)  # For y-axis 1

plt.twinx()  # This creates a new y-axis 
             # for the plots that comes after

plt.plot(x, sin_x, label="sin(x)",
         color="orchid")
plt.legend(loc=(.25, .45), frameon=False)  # For y-axis 2
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt("Complete_TAVG_daily.txt",
                   skiprows=23)
year = data[:,1]
anomaly = data[:, -1]

plt.plot(year, anomaly, alpha=.5)
plt.grid(alpha=.25)
plt.hlines(0, year[0], year[-1], color="grey", linestyle="--", colors="grey")
plt.savefig("/Users/yangxuanxi/Downloads/SP2273/learning-portfolio-yxxpa/plotting/temp.png", dpi=150)
```

## 4.2 Styles

```python
plt.style.use("fivethirtyeight")
data = np.loadtxt("Complete_TAVG_daily.txt",
                   skiprows=23)
year = data[:,1]
anomaly = data[:, -1]

plt.plot(year, anomaly, alpha=.5)
plt.grid(alpha=.25)
plt.hlines(0, year[0], year[-1], color="grey", linestyle="--", colors="grey")

plt.xlabel("Date")
plt.ylabel("Temperature Anomaly")
plt.title("Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)")
```
